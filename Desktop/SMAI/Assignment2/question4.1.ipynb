```python
import math
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt
import pickle
from sklearn.mixture import GaussianMixture
```

```python
import cv2
import os

# Specify the directory containing your PNG images
image_directory = 'KIMIA-dataset'

# List all PNG files in the directory
png_files = [f for f in os.listdir(image_directory) if f.endswith('.png')]

# Load each PNG image and store them in a list
images = []
for png_file in png_files:
    image_path = os.path.join(image_directory, png_file)
    image = cv2.imread(image_path)  # Load image using OpenCV
    images.append(image)
# Now, 'images' is a list containing OpenCV image objects (numpy arrays) of your PNG images
# You can manipulate or process these images as needed
images
```

```python
import matplotlib.pyplot as plt
import math

rows = 3  # Number of rows in the grid
cols = 3  # Number of columns in the grid

num_images = len(images)
# Calculating the number of rows and columns needed to display all images
rows = math.ceil(math.sqrt(num_images))
cols = math.ceil(num_images / rows)
# Create a new figure for displaying the images
plt.figure(figsize=(12, 12))
# Display each image in the grid
for idx, image in enumerate(images):
    plt.subplot(rows, cols, idx + 1)
    plt.imshow(image)
    plt.title('Image {}'.format(idx + 1))
    plt.axis('off')  # Turn off axis numbers and ticks

plt.tight_layout()
plt.show()
```
